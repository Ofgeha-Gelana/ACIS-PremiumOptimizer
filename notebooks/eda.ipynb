```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical columns
numerical_columns = ['TotalPremium', 'TotalClaims']  # Add other numerical columns
df[numerical_columns].hist(bins=20, figsize=(10, 6))
plt.suptitle('Histograms of Numerical Variables')
plt.show()

# Bar charts for categorical columns (e.g., 'ZipCode', 'InsuranceType')
categorical_columns = ['PostalCode', 'CoverCategory']  # Add other categorical columns
for column in categorical_columns:
    sns.countplot(x=df[column])
    plt.title(f'Distribution of {column}')
    plt.show()
```
