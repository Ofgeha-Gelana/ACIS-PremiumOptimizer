#### Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical columns
numerical_columns = ['TotalPremium', 'TotalClaims']  # Add other numerical columns
df[numerical_columns].hist(bins=20, figsize=(10, 6))
plt.suptitle('Histograms of Numerical Variables')
plt.show()

# Bar charts for categorical columns (e.g., 'ZipCode', 'InsuranceType')
categorical_columns = ['PostalCode', 'CoverCategory']  # Add other categorical columns
for column in categorical_columns:
    sns.countplot(x=df[column])
    plt.title(f'Distribution of {column}')
    plt.show()
```

#### Bivariate or Multivariate Analysis

```python
# Scatter plot between TotalPremium and TotalClaim
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['TotalPremium'], y=df['TotalClaims'], hue=df['PostalCode'])
plt.title('Scatter Plot: TotalPremium vs. TotalClaim by ZipCode')
plt.show()



# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# # Correlation matrix
# correlation_matrix = df.corr()
# sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
# plt.title('Correlation Matrix')
# plt.show()
```

####  Data Comparison
##### Trends Over Geography:

```python
# Boxplot for TotalPremium across ZipCode
plt.figure(figsize=(10, 6))
sns.boxplot(x='PostalCode', y='TotalPremium', data=df)
plt.title('TotalPremium across ZipCode')
plt.show()

# Compare InsuranceType across ZipCode
plt.figure(figsize=(10, 6))
sns.countplot(x='PostalCode', hue='CoverCategory', data=df)
plt.title('Insurance Type Distribution by ZipCode')
plt.show()
```

#### Outlier Detection

```python
# Box plot for TotalPremium to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['TotalPremium'])
plt.title('Box Plot: TotalPremium')
plt.show()

# Box plot for TotalClaim to detect outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['TotalClaims'])
plt.title('Box Plot: TotalClaim')
plt.show()
```
